```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
dir(iris_data)
```

```python
feature_names = iris_data.feature_names
feature_names
```

```python
iris_data.data[:3]
```

```python
target_names = iris_data.target_names
target_names
```

```python
target = iris_data.target
target
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figheight(5)
fig.set_figwidth(10)
ax1.scatter(iris_data.data[:,0], iris_data.data[:,1], cmap='Spectral',c=target, edgecolor='k')
ax1.set_xlabel("sepal length (cm)")
ax1.set_ylabel("sepal width (cm)")

ax2.scatter(iris_data.data[:,2], iris_data.data[:,3], cmap='Spectral',c=target, edgecolor='k')
ax2.set_xlabel("petal  length (cm)")
ax2.set_ylabel("petal  width (cm)")
fig.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris_data.data, target, test_size=0.4)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=150)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
prediction_enum = model.predict([[5.1, 3.5, 1.4, 0.2]])
flower_type = target_names[prediction_enum]
flower_type
```

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(8,5))
sn.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
